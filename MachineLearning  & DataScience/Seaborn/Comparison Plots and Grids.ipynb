```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python

sns.jointplot(data=df, x='math score', y='reading score', kind='hex')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', alpha=0.2)
```

```python
sns.jointplot(data=df, x='math score', y='reading score', hue='gender')
```

```python
sns.pairplot(data=df, hue='gender', palette='Dark2')
```

```python
sns.pairplot(data=df, hue='gender', corner=True, palette='Dark2')
```

# Seaborn Grid

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch', row='test preparation course')
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='violin', palette='Dark2', col='lunch', row='test preparation course')
```

```python
sns.pairplot(data=df)
```

```python
g = sns.PairGrid(data=df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```
