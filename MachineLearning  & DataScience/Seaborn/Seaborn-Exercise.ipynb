**График scatterplot пытается показать зависимость между количеством дней на работе (days employed) и возрастом человека (DAYS_BIRTH) для всех, кроме безработных.**

```python
plt.figure(figsize=(10, 8), dpi=200)
sns.scatterplot(data=df.drop(df[df['DAYS_EMPLOYED'] > 0].index) * -1, x='DAYS_BIRTH', y='DAYS_EMPLOYED', 
                alpha=0.01, linewidth=1)
```

```python
plt.figure(figsize=(8, 4))
df_ages = pd.DataFrame(np.array(-1 * df['DAYS_BIRTH'] / 365), columns=['Age in Years'])
sns.histplot(data=df_ages, x='Age in Years', bins=45, color='red', edgecolor='black', linewidth=2)
```

**Этот график отображает информацию только для *половины всех людей*, имеющих доход в этом наборе данных. Причём мы берём "нижнюю" половину - тех, кто зарабатывает меньше. Графики boxplot отображены для каждого значения колонки NAME_FAMILY_STATUS, показывая распределения их общего дохода. Раскраска hue сделана по колонке "FLAG_OWN_REALTY".**

```python
plt.figure(figsize=(12, 8))
df = df.sort_values(by='AMT_INCOME_TOTAL')
sns.boxplot(data=df.head(len(df) // 2), x='NAME_FAMILY_STATUS', y='AMT_INCOME_TOTAL', hue='FLAG_OWN_REALTY')
plt.legend(loc=[1.02, 0.7], title='FLAG_OWN_REALTY')
plt.title('income Totals per Family Status for Bottom Half of Earners')
```

**Эта тепловая карта heatmap показывает корреляцию между колонками в наборе данных. Корреляцию можно посчитать с помощью метода .corr().  Обратите внимание, что колонка FLAG_MOBIL имеет NaN-корреляцию со всеми другими колонками - это видно в результатах метода .corr().**

```python
sns.heatmap(data=df.drop('FLAG_MOBIL', axis=1).corr(numeric_only=True), linewidth=.5, cmap='viridis')
```
