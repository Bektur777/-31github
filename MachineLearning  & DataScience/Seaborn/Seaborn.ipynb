```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

# Scatterplots

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='Dark2', size='work experience', style='level of education')
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, s=200, alpha=0.2)
```

# Distribution Plots

```python
# plt.figure(figsize=(5, 4), dpi=200)
sns.rugplot(x='salary', data=df, height=.5)
```

```python
# sns.set(style='darkgrid')
sns.displot(x='salary', data=df, bins=20, kde=True, rug=True, color='red', edgecolor='blue', linewidth=1)
```

```python
sns.histplot(x='salary', data=df, bins=20)
```

```python
sns.kdeplot(x='salary', data=df)
```

# Categorical Plots

## Categorical statistics

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(data=df, x='level of education')
# plt.ylim(90, 260)
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(data=df, x='level of education', hue='division', palette='Set2')
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, errorbar='sd', hue='division')
plt.legend(loc=(1.02, 0.65))
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
df
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.boxplot(data=df, y='math score', x='test preparation course')
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.boxplot(data=df, y='math score', x='parental level of education', hue='test preparation course')
plt.legend(loc=(1.05, 0.65))
```

```python
plt.figure(figsize=(10, 8), dpi=200)
sns.boxplot(data=df, x='math score', y='parental level of education', hue='test preparation course', palette='Set2')
plt.legend(loc=(1.05, 0.65))
```

```python
plt.figure(figsize=(10, 8), dpi=200)
sns.violinplot(data=df, x='math score', y='parental level of education', hue='test preparation course', 
               palette='Set2', split=True, inner=None)
plt.legend(loc=(1.05, 0.65))
```

```python
plt.figure(figsize=(10, 8), dpi=200)
sns.violinplot(data=df, x='math score', y='parental level of education', split=True, inner='quartile', bw=0.1)
```

```python
plt.figure(figsize=(8, 4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', hue='test preparation course', size=4)
```

```python
plt.figure(figsize=(8, 4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', hue='test preparation course', dodge=True, size=4)
plt.legend(loc=(1.05, 0.65))
```

```python
sns.boxenplot(data=df, x='math score', y='test preparation course', hue='gender')
```
