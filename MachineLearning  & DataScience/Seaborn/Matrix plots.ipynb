```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
```

```python
sns.heatmap(data=df)
```

```python
plt.figure(dpi=200)
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidths=.5, annot=True, cmap='viridis')
```

```python
plt.figure(dpi=200)
sns.clustermap(data=df.drop('Life expectancy', axis=1), linewidths=.5, annot=True, cmap='viridis')
```

```python
plt.figure(dpi=200)
sns.clustermap(data=df.drop('Life expectancy', axis=1), linewidths=.5, annot=True, cmap='viridis', column_cluster=False)
```
